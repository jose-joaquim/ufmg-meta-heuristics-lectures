```python
import polars as pl
import matplotlib.pyplot as plt

filename='rat195'

df = pl.read_csv(f'{filename}.txt')
df.with_columns(Perc_Diff=100*(pl.col('Initial_Population') - pl.col('Final_Population'))/pl.col('Initial_Population')).describe()
```

```python
fig, axs = plt.subplots(1, 1)
fig.tight_layout()
axs.boxplot([df['Initial_Population'].to_list(), df['Final_Population'].to_list()], labels=['Initial Population', 'Final Population'])
axs.set_ylabel('Valor da função objetivo')

plt.savefig(f'{filename}.eps', bbox_inches = 'tight')
```

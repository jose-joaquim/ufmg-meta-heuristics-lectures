```python
import polars as pl
import matplotlib.pyplot as plt


df = pl.read_csv('data.txt',separator=" ", schema={'Initial Population' : pl.Float64, 'Final Population' : pl.Float64})
df.describe()

fig, axs = plt.subplots(1, 1)
fig.tight_layout()
axs.boxplot([df['Initial Population'].to_list(), df['Final Population'].to_list()], labels=['Initial Population', 'Final Population'])
axs.set_ylabel('Valor da função objetivo')

plt.savefig('boxplot.eps', bbox_inches = 'tight')
```
